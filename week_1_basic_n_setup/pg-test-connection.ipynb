### 1. sqlalchemy – Thư viện ORM (Object-Relational Mapping)
SQLAlchemy cho phép bạn làm việc với cơ sở dữ liệu bằng Python code thay vì viết SQL thủ công.

Nó cung cấp một lớp trừu tượng cho các câu truy vấn SQL.

Hỗ trợ rất nhiều hệ quản trị cơ sở dữ liệu: PostgreSQL, MySQL, SQLite, SQL Server, v.v

### psycopg2-binary – PostgreSQL driver cho Python
Đây là driver (trình điều khiển) để Python có thể kết nối và tương tác trực tiếp với cơ sở dữ liệu PostgreSQL.

sqlalchemy sẽ dùng driver này khi làm việc với PostgreSQL.

````python
from sqlalchemy import create_engine
````

````python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')

# engine = create_engine('postgresql://<username>:<password>@<host>:<port>/<database>')
````

If you cannot connect to ny_taxi, use `postgres` database instead

```python
engine = create_engine('postgresql://root:root@localhost:5432/postgres')
```

````python
engine.connect()
````

````python
query = """
SELECT 1 as number;
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT *
FROM pg_catalog.pg_tables
WHERE schemaname != 'pg_catalog' AND 
    schemaname != 'information_schema';
"""

pd.read_sql(query, con=engine)
````
